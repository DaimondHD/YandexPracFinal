```python
query = '''
SELECT count(book_id)
FROM books
where publication_date >= '2000-01-02'
'''
pd.io.sql.read_sql(query, con = engine)
```

Всего 819 книг

Для каждой книги посчитайте количество обзоров и среднюю оценку;

```python
query = '''
WITH  rev as(SELECT title,books.book_id ,count(text)
FROM books
left join reviews  on books.book_id=reviews.book_id 
group by books.book_id),

rat as (SELECT title,books.book_id ,avg(rating)
FROM books
left join ratings  on books.book_id=ratings.book_id
group by books.book_id)

SELECT  rev.title,rev.book_id ,count,avg
from rev
left  join rat on rev.book_id=rat.book_id
order by count DESC,avg DESC

'''
pd.io.sql.read_sql(query, con = engine)
 
```

Лидируют Twilight (Twilight #1), The Hobbit or There and Back Again	и The Catcher in the Rye.

Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```python
query = '''
SELECT publishers.publisher_id ,publisher,count(book_id) AS count
FROM publishers
inner join books  on publishers.publisher_id=books.publisher_id 
where num_pages >50
group by publishers.publisher_id
order by count desc
'''
pd.io.sql.read_sql(query, con = engine)
```

Penguin Books - явный лидер. Издательство выпустило 42 книги.

Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```python
query = '''
SELECT  author,avg(avg)
from (SELECT  author,title,books.book_id,avg(rating),count(ratings.rating_id)
FROM authors
inner join books  on authors.author_id  =books.author_id   
inner join ratings  on books.book_id =ratings.book_id  
group by author,books.book_id
having count(ratings.rating_id)>50
order by avg(rating) desc) as avgg
group by author
order by avg desc
'''
pd.io.sql.read_sql(query, con = engine)
```

Самая высокая средняя оценка - 4.283844 у J.K. Rowling/Mary GrandPré.

Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query = '''
WITH rat as (SELECT ratings.username,count(rating)
FROM ratings
group by ratings.username
having count(rating)>50
order by count desc),

rev as (SELECT username, COUNT(text) as text_count
FROM reviews
group by username)

select avg(text_count) from rev
inner join rat  using(username)



'''
pd.io.sql.read_sql(query, con = engine)
```

Получили всего 24.3 - среднее количество обзоров от пользователей, которые поставили больше 50 оценок.
